```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Display the first few values and statistics
#print("First 10 values of the distribution:")
#print(s[:10])
#print(f"Mean: {np.mean(s):.4f}")
#print(f"Standard Deviation: {np.std(s):.4f}")
#plt.figure(figsize=(10, 6))
plt.hist(s, bins=50, density=True, alpha=0.6, color='g')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of a Normal Distribution')
plt.grid(True)
plt.show()
```
