```python
 import numpy as np

def calculate_statistics():
    while True:
        user_input = input("Enter your dataset as a comma-separated list of numbers (e.g., 1,2,3,4,5): ")
        try:

            data = [float(x.strip()) for x in user_input.split(',')]
            if not data: # Check if the list is empty after conversion
                print("No valid numbers entered. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please make sure all values are numbers and separated by commas.")


    data_array = np.array(data)


    mean_val = np.mean(data_array)
    median_val = np.median(data_array)
    variance_val = np.var(data_array)
    std_dev_val = np.std(data_array)


    print(f"\nDataset: {data}")
    print(f"Mean: {mean_val:.2f}")
    print(f"Median: {median_val:.2f}")
    print(f"Variance: {variance_val:.2f}")
    print(f"Standard Deviation: {std_dev_val:.2f}")


calculate_statistics()
```
