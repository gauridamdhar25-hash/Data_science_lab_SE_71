```python
 # Sample dataset
data = [10, 20, 30, 40, 50]

# --- Mean Calculation ---
sum_data = sum(data)
count_data = len(data)

mean = sum_data / count_data
print(f"Dataset: {data}")
print(f"Mean: {mean}")

# --- Median Calculation ---
sorted_data = sorted(data)

if count_data % 2 == 1:
    median = sorted_data[count_data // 2]
else:
    mid1 = sorted_data[count_data // 2 - 1]
    mid2 = sorted_data[count_data // 2]
    median = (mid1 + mid2) / 2

print(f"Median: {median}")

# --- Variance Calculation ---
sum_squared_diff = 0
for x in data:
    diff = x - mean
    sum_squared_diff += diff ** 2

# Population variance
variance = sum_squared_diff / count_data
print(f"Variance: {variance}")

# Standard Deviation
standard_deviation = variance ** 0.5
print(f"Standard Deviation: {standard_deviation}")
```
