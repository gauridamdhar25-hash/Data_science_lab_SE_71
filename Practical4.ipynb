```python
print("Generate a normal distribution and plot its histogram.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define parameters
mean = 8
std_dev=1
size = 1000
#Generate normal distribution
data = np.random.normal(mean, std_dev, size)
#Plot histogram
plt.hist(data, bins=30, color='blue', edgecolor='black')
plt.title("Normal Distribution Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Compute mean, median, variance, and standard deviation for a dataset.")
```

```python
# Input dataset
data=[10, 20, 30, 40, 50]
print("Dataset:", data)
# Number of elements
n=len(data)
#Mean
sum_data = sum(data)
mean = sum_data /n
print("Mean =", mean)
# Median
data.sort()
if n % 2 == 0:
  median = (data[n//21] + data[n//2]) / 2
else:
  median = data[n//2]
print("Median -", median)
# Variance
variance_sum= 0
for i in data:
  variance_sum = variance_sum + (i- mean) ** 2
variance= variance_sum/n
print("Variance =", variance)
# Standard Deviation
std_dev= variance ** 0.5
print("Standard Deviation =", std_dev)
```

```python
print("Simulate a coin toss experiment and compute probabilities.")
```

```python
import random
# take input from user
n = int(input("Enter number of coin tosses: "))
heads = 0
tails = 0
#simulate coin toss
for i in range(n):
    toss = random.choice(["H", "T"])
    if toss == "H":
       heads += 1
    else:
       tails += 1
# compute probabilities
p_heads = heads /n
p_tails = tails /n
print("Number of Heads:", heads)
print("Number of Tails:", tails)
print("Probability of Heads:", p_heads)
print("Probability of Tails:", p_tails)
```
