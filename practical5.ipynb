```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/bird.csv")
print(df.head())
```

```python
filtered_df = df[df["huml"] > 70]
print(filtered_df)

filtered_df2 = df[(df["huml"] > 70) & (df["feml"] > 40)]
print(filtered_df2)
```

```python
grouped_mean = df.groupby("type").mean(numeric_only=True)
print(grouped_mean)
```
