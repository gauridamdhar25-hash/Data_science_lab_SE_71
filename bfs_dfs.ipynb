```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    visited.add(start_node)

    bfs_path = []

    while queue:
        current_node = queue.popleft()
        bfs_path.append(current_node)

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return bfs_path


graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("BFS Traversal starting from node 'A':")
print(bfs(graph, 'A'))

print("\nBFS Traversal starting from node 'C':")
print(bfs(graph, 'C'))
```

```python
def dfs(graph, start_node, visited=None, dfs_path=None):
    if visited is None:
        visited = set()
    if dfs_path is None:
        dfs_path = []

    visited.add(start_node)
    dfs_path.append(start_node)

    for neighbor in graph[start_node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited, dfs_path)

    return dfs_path


graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("DFS Traversal starting from node 'A':")
print(dfs(graph, 'A'))

print("\nDFS Traversal starting from node 'C':")
print(dfs(graph, 'C'))
```
