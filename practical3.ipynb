```python
print("Create a 3x3 square matrix and compute its determinant.")
```

```python
import numpy as np
A = np.array([[2,3],
              [4,5]])
B = np.array([10,20])

x,y = np.linalg.solve(A,B)
print(f"the solution is: x = {x:.2f}, y = {y:.2f}")
```

```python
print("Find the inverse of the matrix.")
```

```python
import numpy as np
matrix = np.array([
    [1,2,3],
    [0,1,4],
    [5,6,0]
])

print("Original 3*3 matrix: ")
print(matrix)

try:
  inverse_matrix = np.linalg.inv(matrix)
  print("\nInverse of the matrix:")
  print(inverse_matrix)
except np.linalg.LinAlgError:
    print("\nError: The matrix is singular and does not have an inverse.")
```

```python
print("Solve the following system of linear equations using NumPy: 2x+3y=10, 4x+5y=20")
```

```python
matrix = np.array([[1,2,3],
                   [4,5,6],
                   [4,8,9]])

print("The 3*3 matrix is: ")
print(matrix)

determinant = np.linalg.det(matrix)

print(f"\nThe determinant of the matrix is: {determinant}")
```
