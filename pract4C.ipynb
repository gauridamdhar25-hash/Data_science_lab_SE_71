```python
import random

tosses = int(input("Enter number of tosses: "))
heads = 0
tails = 0

for i in range(tosses):
    result = random.choice(["H", "T"])
    if result == "H":
        heads += 1
    else:
        tails += 1

prob_heads = heads / tosses
prob_tails = tails / tosses

print("Total Tosses:", tosses)
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```
