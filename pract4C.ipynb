```python
import random

def coin_toss_simulation():
    try:
        num_tosses = int(input("Enter the number of coin tosses to simulate: "))
        if num_tosses <= 0:
            print("Please enter a positive integer for the number of tosses.")
            return
    except ValueError:
        print("Invalid input. Please enter an integer.")
        return

    heads_count = 0
    tails_count = 0

    print(f"\nSimulating {num_tosses} coin tosses...")
    for _ in range(num_tosses):
        result = random.choice(["Heads", "Tails"])
        if result == "Heads":
            heads_count += 1
        else:
            tails_count += 1

    print("\n--- Simulation Results ---")
    print(f"Total tosses: {num_tosses}")
    print(f"Heads count: {heads_count}")
    print(f"Tails count: {tails_count}")

    # Calculate probabilities
    prob_heads = (heads_count / num_tosses) * 100 if num_tosses > 0 else 0
    prob_tails = (tails_count / num_tosses) * 100 if num_tosses > 0 else 0

    print("\n--- Probabilities ---")
    print(f"Probability of Heads: {prob_heads:.2f}%")
    print(f"Probability of Tails: {prob_tails:.2f}%")

# Call the function to run the simulation
coin_toss_simulation()
```
