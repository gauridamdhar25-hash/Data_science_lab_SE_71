```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    traversal_order = []

    visited.add(start_node)

    while queue:
        current_node = queue.popleft()
        traversal_order.append(current_node)

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return traversal_order

# Example Graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': [ 'D' ],
    'C': [ 'E'],
    'D': ['F'],
    'E': [],
    'F': []
}

print("BFS Traversal starting from 'A':", bfs(graph, 'A'))
```

```python
def dfs(graph, start_node, visited=None):
    if visited is None:
        visited = set()

    traversal_order = []

    visited.add(start_node)
    traversal_order.append(start_node)

    for neighbor in graph[start_node]:
        if neighbor not in visited:
            traversal_order.extend(dfs(graph, neighbor, visited))

    return traversal_order

# Example Graph (Adjacency List) - same as BFS example
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("DFS Traversal starting from 'A':", dfs(graph, 'A'))
```

```python
import heapq

def a_star_custom_graph(graph, heuristics, start, goal):
    """
    Implements the A* pathfinding algorithm for a custom graph.

    Args:
        graph (dict): Adjacency list representation of the graph.
                      Keys are nodes, values are lists of (neighbor, cost) tuples.
        heuristics (dict): Dictionary mapping each node to its heuristic cost
                           (estimated cost from node to goal).
        start: The starting node.
        goal: The goal node.

    Returns:
        list or None: The shortest path as a list of nodes,
                      or None if no path is found.
    """

    # Priority queue to store (f_cost, g_cost, current_node, path)
    # f_cost = g_cost + h_cost
    open_set = [(heuristics[start], 0, start, [start])]
    heapq.heapify(open_set)

    # Dictionary to store the g_cost of the cheapest path found to a node
    g_costs = {start: 0}

    # Set to keep track of visited nodes
    closed_set = set()

    while open_set:
        f_cost, g_cost, current_node, path = heapq.heappop(open_set)

        if current_node == goal:
            return path

        if current_node in closed_set:
            continue

        closed_set.add(current_node)

        for neighbor, cost in graph.get(current_node, []):
            new_g_cost = g_cost + cost

            if neighbor not in g_costs or new_g_cost < g_costs[neighbor]:
                g_costs[neighbor] = new_g_cost
                h_cost = heuristics.get(neighbor, float('inf')) # Use infinity if heuristic not defined
                new_f_cost = new_g_cost + h_cost
                new_path = path + [neighbor]
                heapq.heappush(open_set, (new_f_cost, new_g_cost, neighbor, new_path))

    return None  # No path found

# Define the graph based on your description:
# a is root node
# b and c are children of a
# d and e are child node of b
# g is child node of e
# f is child node of c
graph = {
    'a': [('b', 1), ('c', 1)], # Assuming cost of 1 for each edge
    'b': [('d', 1), ('e', 1)],
    'c': [('f', 1)],
    'd': [],
    'e': [('g', 1)],
    'f': [],
    'g': []
}

# Define heuristic values for each node to the goal 'g'.
# These are arbitrary estimates for demonstration purposes.
# For a true A* search, these should be carefully chosen admissible heuristics.
heuristics = {
    'a': 8,  # Estimated distance from 'a' to 'g'
    'b': 7,  # Estimated distance from 'b' to 'g'
    'c': 5,  # 'c' cannot reach 'g' in this graph structure, so a high value
    'd':9 ,  # 'd' cannot reach 'g', high value
    'e': 2,  # Estimated distance from 'e' to 'g'
    'f': 3,  # 'f' cannot reach 'g', high value
    'g': 10   # Goal node has 0 heuristic cost
}

start_node = 'a'
goal_node = 'g'

path = a_star_custom_graph(graph, heuristics, start_node, goal_node)

if path:
    print(f"Path found from {start_node} to {goal_node}: {path}")
else:
    print(f"No path found from {start_node} to {goal_node}.")
```

```python
class Node:
    def __init__(self, name, value=None, children=None):
        self.name = name
        self.value = value  # Only for leaf nodes
        self.children = children if children is not None else []

    def is_leaf(self):
        return not self.children

    def __repr__(self):
        return f"Node('{self.name}', value={self.value})"

def minimax(node, maximizing_player):
    if node.is_leaf():
        return node.value

    if maximizing_player:
        max_eval = -float('inf')
        for child in node.children:
            eval = minimax(child, False)
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for child in node.children:
            eval = minimax(child, True)
            min_eval = min(min_eval, eval)
        return min_eval

# Define the graph structure with placeholder leaf values
# Leaves
D = Node('D', value=4)
E = Node('E', value=7)
F = Node('F', value=2)
G = Node('G', value=9)
H = Node('H', value=1)
I = Node('I', value=5)

# Intermediate nodes
B = Node('B', children=[D, E, F])
C = Node('C', children=[G, H, I])

# Root node
A = Node('A', children=[B, C])

print("Graph Structure (from root A):")
print(f"  Children of A: {A.children}")
print(f"    Children of B: {B.children}")
print(f"    Children of C: {C.children}")

# Run Minimax algorithm, assuming the root player (A) is the maximizing player
optimal_value = minimax(A, True)
print(f"\nThe optimal value for the root node A (maximizing player) is: {optimal_value}")
```
